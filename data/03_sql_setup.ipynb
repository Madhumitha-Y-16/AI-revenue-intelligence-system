```python
# connecting python to postgresql
from sqlalchemy import create_engine
import pandas as pd

engine = create_engine('postgresql://postgres:Madhumitha05@localhost:5432/salesforce_pro_db')

print('connected successfully!')
```

```python
# loading the cleaned csv into postgresql
df = pd.read_csv('salesforce_pro_cleaned.csv')

# pushing it to postgresql , this creates the table automatically
df.to_sql('sales_data', engine, if_exists='replace', index=False)

print(f'data loaded successfully — {len(df)} rows pushed to postgresql')
```

```python
# checking if the data made it to postgresql
result = pd.read_sql('SELECT COUNT(*) as total_rows FROM sales_data', engine)
print(result)
```
